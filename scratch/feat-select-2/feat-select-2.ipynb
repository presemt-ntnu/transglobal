## Feature selection through frequency thresholding

Feature selection with different frequency cut-offs,
with Naive Bayes models with class priors from corpus,
with MFT vectors.

1. Filter out features with a minimum nount below threshold:
   min_count = [1, 5, 10, 25, 50, 100, 250, 500]
   
2. Filter out features with a document/instance frequence above threshold:
   max_freq = [1.0, 0.5, 0.1, 0.075, 0.05, 0.025, 0.01, 0.005]

```python
analyse_fs_2.analyse_fs_2("accuracy")
```

Note that for some combinations, no features at all are selected, so there is no model for the word. 
As a result the number of "ignored" in accuracy calculations may vary.
For some datapoints, the accuracy score may therefore be based few data points, perhaps only a couple (cf. Greek).

It seems it is generally save to have min_count=10. Lower or higher values may slightly lowers scores.

max_freq seems not to matter a lot, except in combination with high values of min_count. However, this may be due to the fact that with thise values a lot of models are skipped because no features weree selected. Max_freq=0.05 seems OK.

```python
analyse_fs_2.analyse_fs_2("bleu")
```

Effect of frequency filtering on Bleu scores is marginal at least.

## Conclusion

It seems the effect of frequency filtering is rather small. Main advantage may be smaller/faster models.

min_count=10 and max_freq=0.05 seem reasonable default values.
