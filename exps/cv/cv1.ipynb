# CV1: Cross-Validated Results on Context Samples

Language pair is German-English (de-en). Input consists of an ambiguity table containing 5408 ambiguous German lempos and their candidate translations. We collect context samples (max 10k per translation candidate) and use them to train a Multinomial Naive Bayes classifier. NB is fast and seems to do better than e.g. Perceptron or NearestCentroid. Linear SVM seems to do better, but is slow. Context words are limited to those occurring in the translation dictionary; no further pruning or feature selection (chi2) is carried out, because this seemed not to improve scores. Evaluation through 3-fold stratified cross validation. Precisio, recall and F1 are macro-averaged scores, i.e. not taking skewed class distributions into account. This gives an indication of the power of classifiers to disambiguate when the context is pure, that is, not polluted by the word-for-word translations in the normal WTD experiments.

## Preliminaries

The first step consist of creating the data sets and cross-validating a classifier, for every ambiguous source lempos. ## Number of candidate translations

The minimum number of translation candidates is 2; the maximum number is 50, showing roughly a Zipfian distribution.

```python
cv1.plot_2(results)
```

It is to be expected that scores go down when the number of candidate translations increases. The following plot confirms that the range of F1 scores becomes more narrow when there are more translations candidates (= classes).

```python
cv1.plot_1(results)
```

## Part-of-Speech

POS is dominated by content words (n, adj, adv, v*.full). F1 scores are best for nouns and adjectives, reasonably good for content verbs and verbal particles, but less so for adverbs. Scores are worst for possesive pronouns and cardinals and articles. Function words are probably best diambiguated on the basis of local rather than global context. 

```python
cv1.pos(results)
```
